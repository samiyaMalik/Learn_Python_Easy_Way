## 2Step: Handling Missing Values:

```python
df.isnull()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

### Visualization NAN  values

```python
sns.heatmap(df.isnull(), yticklabels=False,cbar=False,cmap='viridis')
```

```python
# Fill missing values with a specific value
df['age'] = df.age.fillna(value=df['age'].mean())
df
```

```python
df['embarked'] = df.embarked.fillna(value=df['embarked'].mode()[0])
df
```

```python
df['embark_town'] = df.embark_town.fillna(value =df['embark_town'].mode()[0])
df
```

```python
df.isnull().sum()*100/len(df)
```

# how to add a columns and make a age group

```python
df['age_group']= pd.cut(df['age'],[0,10,18,30,50,80],
labels=['kids','Child','Youth','Middle','Old'])
df
```

```python
df['Family_size'] = pd.cut(df['sibsp'], bins=[0,4,6,8], 
labels=['small', 'medium', 'Large'])
df
```

```python
df.isnull().sum()
```

```python
df['Family_size'] = df.Family_size.fillna(value =df['Family_size'].mode()[0])
df
```

```python
df['age_group'] = df.age_group.fillna(value =df['age_group'].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns='deck',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
import numpy as np
kashti['age'] = kashti['age'].fillna(kashti['age'].median())
kashti['fare'] = kashti['fare'].fillna(kashti['fare'].median())
#log tramsformation
kashti['age']=np.log(kashti['age'])
kashti['fare']=np.log(kashti['fare'])
```
