# Names of the Distributions

    1. Normal Distribution
    2. Uniform Distribution
    3. Binomial Distribution
    4. Poisson Distribution
    5. Exponential Distribution
    6. Gamma Distribution
    7. Beta Distribution
    8. Lognormal Distribution
    9. Chi-Square Distribution
    10. Student's t-Distribution
    11. F-Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, beta, lognorm, chi2, t, f
from scipy.stats import norm, uniform, binom, expon
```

```python
# Normal Distribution
mu = 0
sigma = 1
x_normal = np.linspace(-5, 5, 1000)
y_normal = norm.pdf(x_normal, mu, sigma)

plt.figure(figsize=(10, 6))
plt.plot(x_normal, y_normal, label='Normal Distribution')
plt.title('Normal Distribution (=0, =1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Uniform Distribution
x_uniform = np.linspace(0, 1, 1000)
y_uniform = uniform.pdf(x_uniform)

plt.figure(figsize=(10, 6))
plt.plot(x_uniform, y_uniform, label='Uniform Distribution')
plt.title('Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Binomial Distribution
n = 10
p = 0.5
x_binomial = np.arange(0, n+1)
y_binomial = binom.pmf(x_binomial, n, p)

plt.figure(figsize=(10, 6))
plt.stem(x_binomial, y_binomial, label='Binomial Distribution')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('x')
plt.ylabel('Probability Mass')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Exponential Distribution
scale_parameter = 1 / 5  # Mean = 1 / 
x_exponential = np.linspace(0, 20, 1000)
y_exponential = expon.pdf(x_exponential, scale=scale_parameter)

plt.figure(figsize=(10, 6))
plt.plot(x_exponential, y_exponential, label='Exponential Distribution')
plt.title('Exponential Distribution (=5)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gamma Distribution
k_gamma = 2
theta_gamma = 1 / 0.5  # rate = 1 / scale
x_gamma = np.linspace(0, 30, 1000)
y_gamma = gamma.pdf(x_gamma, k_gamma, scale=theta_gamma)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_gamma, y_gamma, label='Gamma Distribution')
plt.title('Gamma Distribution (k=2, =2)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Beta Distribution
alpha_beta = 2
beta_beta = 2
x_beta = np.linspace(0, 1, 1000)
y_beta = beta.pdf(x_beta, alpha_beta, beta_beta)

plt.figure(figsize=(10, 6))
plt.plot(x_beta, y_beta, label='Beta Distribution')
plt.title('Beta Distribution (=2, =2)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Lognormal Distribution
mu_lognorm = 0
sigma_lognorm = 1
x_lognorm = np.linspace(0, 5, 1000)
y_lognorm = lognorm.pdf(x_lognorm, sigma_lognorm, scale=np.exp(mu_lognorm))

plt.figure(figsize=(10, 6))
plt.plot(x_lognorm, y_lognorm, label='Lognormal Distribution')
plt.title('Lognormal Distribution (=0, =1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Chi-Square Distribution
df_chi2 = 5
x_chi2 = np.linspace(0, 20, 1000)
y_chi2 = chi2.pdf(x_chi2, df_chi2)

plt.figure(figsize=(10, 6))
plt.plot(x_chi2, y_chi2, label='Chi-Square Distribution')
plt.title('Chi-Square Distribution (df=5)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Student's t-Distribution
df_t = 10
x_t = np.linspace(-5, 5, 1000)
y_t = t.pdf(x_t, df_t)

plt.figure(figsize=(10, 6))
plt.plot(x_t, y_t, label="Student's t-Distribution")
plt.title("Student's t-Distribution (df=10)")
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# F-Distribution
dfn_f = 5
dfd_f = 10
x_f = np.linspace(0, 5, 1000)
y_f = f.pdf(x_f, dfn_f, dfd_f)

plt.figure(figsize=(10, 6))
plt.plot(x_f, y_f, label='F-Distribution')
plt.title('F-Distribution (dfn=5, dfd=10)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
